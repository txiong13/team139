```python
import pandas as pd
df = pd.read_csv("facebook_combined.txt", sep=" ")
nodes = set(df["0"].unique().tolist()).union(set(df["1"].unique().tolist()))
```

```python
import networkx as nx
import collections 
```

```python
G = nx.Graph()
for node in nodes:
    G.add_node(node)
nx.set_node_attributes(G, "status", "S")
for index, row in df.iterrows():
    G.add_edge(row["0"], row["1"])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(30,30)) 
nx.draw(G)
```

```python
graph = nx.to_dict_of_dicts(G)

import pickle
f = open("fb_graph.pkl","wb")
pickle.dump(graph,f)
f.close()

# To import graph, run:
# with open('fb_graph.pkl', 'rb') as f:
#     return pickle.load(f)
# G = nx.from_dict_of_dicts(G)
```
