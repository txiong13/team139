```python
import pickle
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
import random
import scipy
from scipy.stats import lognorm
from scipy.stats import norm
import math
#import seaborn as sns
from random import sample
import pandas 
import src.test_strategies
```

```python
no_test_df = pandas.read_csv("simulations/no_test.csv", index_col = 0)
connect_test_df = pandas.read_csv("simulations/connect_norm_test.csv")
random_test_df = pandas.read_csv("simulations/random_test.csv")
infected_test_df = pandas.read_csv("simulations/infected_test.csv")
```

```python
plt.plot([i for i in range(50)], no_test_df.mean(axis = 1), label = "Untested", color = "cornflowerblue")
plt.fill_between([i for i in range(50)], no_test_df.mean(axis = 1) + 2*no_test_df.std(axis = 1), no_test_df.mean(axis = 1) - 2*no_test_df.std(axis = 1), color = "cornflowerblue", alpha = .1)

plt.plot([i for i in range(50)], connect_test_df.mean(axis = 1), label = "Test highly connected", color = "darkorange")
plt.fill_between([i for i in range(50)], connect_test_df.mean(axis = 1) + 2*connect_test_df.std(axis = 1), connect_test_df.mean(axis = 1) - 2*connect_test_df.std(axis = 1), color = "darkorange", alpha = .1)

plt.plot([i for i in range(50)], random_test_df.mean(axis = 1), label = "Test Random", color = "red")
plt.fill_between([i for i in range(50)], random_test_df.mean(axis = 1) + 2*random_test_df.std(axis = 1), random_test_df.mean(axis = 1) - 2*random_test_df.std(axis = 1), color = "red", alpha = .1)

plt.plot([i for i in range(50)], infected_test_df.mean(axis = 1), label = "Test Infected", color = "lightgreen")
plt.fill_between([i for i in range(50)], infected_test_df.mean(axis = 1) + 2*infected_test_df.std(axis = 1), infected_test_df.mean(axis = 1) - 2*infected_test_df.std(axis = 1), color = "lightgreen", alpha = .1)

plt.legend()
plt.show()
```

```python
plt.plot([i for i in range(50)], no_test_df.mean(axis = 1).divide(no_test_df.mean(axis =1)), label = "Untested")
plt.plot([i for i in range(50)], connect_test_df.mean(axis = 1).divide(no_test_df.mean(axis =1)), label = "Test Highly Connected")
plt.plot([i for i in range(50)], random_test_df.mean(axis = 1).divide(no_test_df.mean(axis =1)), label = "Test Random")
plt.plot([i for i in range(50)], infected_test_df.mean(axis = 1).divide(no_test_df.mean(axis =1)), label = "Test Infected")

plt.legend()
plt.show()
```
